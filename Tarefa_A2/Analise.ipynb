# Indicadores Educacionais
O objetivo aqui é apenas desmonstar com dados a diferença de ensino entre diferentes áreas e/ou escolas pelo Brasil. Este já é um problema reconhecido, basta ter conhecimento do dia a dia, mas aqui quero tirar métricas para comprovar essa hipótese.

Dito isso quero comparar os resultados encontrados entre o tipo de rede das escolas e, fazendo uma análise temporal, avaliar se essa desigualdade está diminuindo. Também tomarei como foco especial o ensino fundamental e médio.

```python
# Bibliotecas utilizdas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("br_inep_indicadores_educacionais_brasil.csv")
# Quero ter uma ideia de quantos dados estou trabalhando aqui
df.shape
```

## Explicação das colunas
*   atu: média de alunos por turma
*   had: média de horas-aula diária
*   tdi: taxa de distorção idade-série (quantos % está pelo menos 2 anos atrasado)
*   dsu: percentual de docentes com ensino superior

```python
# Não foi explicado o que as redes publica e total significam, ainda assim tratarei eles como uma rede separada
df.rede.unique()
```

## Análise por rede escolar

```python
# Separando os dados por rede e agrupando por ano
df_fed = df[df["rede"] == "federal"].groupby("ano").mean(numeric_only=True).reset_index()
df_priv = df[df["rede"] == "privada"].groupby("ano").mean(numeric_only=True).reset_index()
df_mun = df[df["rede"] == "municipal"].groupby("ano").mean(numeric_only=True).reset_index()
df_pub = df[df["rede"] == "publica"].groupby("ano").mean(numeric_only=True).reset_index()
df_tot = df[df["rede"] == "total"].groupby("ano").mean(numeric_only=True).reset_index()
df_est = df[df["rede"] == "estadual"].groupby("ano").mean(numeric_only=True).reset_index()

# Agrupando cada df com a sua rede (facilita na hora de plotar)
data = [df_fed, df_priv, df_mun, df_pub, df_tot, df_est]
rede = df.rede.unique()
data = list(zip(data, rede))
```

```python
def plot_cols(data, colunas):
    """
    Gera um gráfico de linhas (lineplot) para uma lista de DataFrames e colunas específicas.

    Entrada:
        - df_list (list): Lista de DataFrames.
        - columns (list): Lista de nomes de colunas para plotar.

    Saída:
        Um gráfico de linhas combinando as colunas fornecidas em cada DataFrame.
    """
    
    plt.figure(figsize=(10, 6))
    
    # Iterar sobre os dados
    for df, r in data:
        for col in colunas:
            if col in df.columns: # Para evitar erros
                sns.lineplot(data=df, x='ano', y=col, label=f'{r}')
    
    
    plt.title(f"ano x {colunas[0]}", fontsize=14)
    plt.xlabel('Ano', fontsize=12)
    plt.ylabel(colunas[0], fontsize=12)
    plt.legend(loc='best')
    plt.grid(True)
```

```python
colunas_interesse_ef = ["ano", "taxa_aprovacao_ef", "taxa_reprovacao_ef", "taxa_abandono_ef", "atu_ef", "had_ef", "tdi_ef", "dsu_ef"]
```
